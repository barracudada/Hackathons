```python
def get_shares_history(board, start_dt, end_dt):
    url1 = 'https://iss.moex.com/iss/history/engines/stock/markets/shares/boards/'
    #board = 'tqbr'
    url2 = '/securities.json?date='
    
    date_range = [i.date().strftime('%Y-%m-%d') for i in list(pd.date_range(start=start_dt, end=end_dt))]
    
    itg = []
    
    for i in date_range:
        url = url1 + board + url2 + i
        itg.append(get_json(url))
        
    return itg

get_shares_history('tqbr', '2023-11-20', '2023-11-24')
```

```python
def get_share_history(ticker):
    url1 = 'https://iss.moex.com/iss/history/engines/stock/markets/shares/boards/TQBR/securities/'
    #ticker = 'sber'
    url2 = '/aggregates.json?'
    url3 = 'from=2023-01-01&till=2023-11-21&interval=24&start=0'

    x= get_json(url1 + url2 + ticker + url3)

    return pd.DataFrame(x['history']['data'], columns=x['history']['columns']) 

get_share_history('sber')
```

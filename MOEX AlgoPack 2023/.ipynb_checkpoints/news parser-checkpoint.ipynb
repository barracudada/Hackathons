```python
def get_attr_from_html(path="example.html"):
    f=codecs.open(path, 'r', 'utf-8')
    document= BeautifulSoup(f.read())
    
    timetag = document.find_all(property='article:published_time')[0]['content']
    dt, tm = timetag.split('T')[0] , timetag.split('T')[1].split('+')[0]
    
    title = document.find_all('title')
    title = title[0].get_text()
    
    text = document.find_all('p', class_='doc__text')

    text_itg = []
    for i in text:
        text_itg.append(i.get_text())

    text_itg = text_itg[:-1]
    text_itg = ' '.join(text_itg)
    
    return ([dt, tm, str(title.encode('utf-8').decode('utf8'))
             , str(text_itg.encode('utf-8').decode('utf8'))])
```

```python
get_attr_from_html(path="example.html")
```

```python
import pandas as pd
import csv
```

```python
with open('news.csv', 'w', encoding='utf-8', newline='') as f:

    data_itg = []

    #list_of_paths = ['example1.html', 'example2.html', ...]
    list_of_paths = ['example.html']
    
    header = ['date', 'time', 'title', 'text']
    writer = csv.writer(f)
    writer.writerow(header)


    for i in list_of_paths:
        data = (get_attr_from_html(path=i))
        writer.writerow(data)
        
        data_itg.append(data)
    
```

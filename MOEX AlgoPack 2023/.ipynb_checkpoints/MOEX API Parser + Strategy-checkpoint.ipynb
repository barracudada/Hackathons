##### Достать исторические значения торгов (все акции в день торгов)

```python
def get_all_shares_history(board='tqbr', start='2023-01-01', end='2023-11-23'):
    """
    'BOARDID',
 'TRADEDATE',
 'SHORTNAME',
 'SECID',
 'NUMTRADES',
 'VALUE',
 'OPEN',
 'LOW',
 'HIGH',
 'LEGALCLOSEPRICE',
 'WAPRICE',
 'CLOSE',
 'VOLUME',
 'MARKETPRICE2',
 'MARKETPRICE3',
 'ADMITTEDQUOTE',
 'MP2VALTRD',
 'MARKETPRICE3TRADESVALUE',
 'ADMITTEDVALUE',
 'WAVAL',
 'TRADINGSESSION',
 'CURRENCYID',
 'TRENDCLSPR'
    """

    x = pd.DataFrame()
    url = 'https://iss.moex.com/iss/history/engines/stock/markets/shares/boards/'+board+'/securities.json?date='

    for i in [i.date().strftime('%Y-%m-%d') for i in list(pd.date_range(start=start, end=end))]:

        y = get_json(url + i)['history']['data']

        x = pd.concat([x, pd.DataFrame(y['securities']['data']
                                       , columns=y['securities']['columns'])]
                                         , axis=0 )
    return x 
```

##### Strategy

```python
class MeanReversion:
    def __init__(self, window_size, fees):
        self.window_size = window_size
        self.fees = fees
        self.prev_prices = []

    def get_update(self, update):
        # по хорошему тот должен быть isinstance(update, CandleUpdate)
        if update['price'] is None:
            return 0
        
        price = update['price']
        if len(self.prev_prices) < self.window_size:
            self.prev_prices.append(price)
            return 0
        else:
            self.prev_prices.append(price)
            self.prev_prices.pop(0)
            if price > np.mean(self.prev_prices) * (1 + self.fees): #  
                return -1
            elif price < np.mean(self.prev_prices) * (1 - self.fees):
                return 1
            else:
                return 0
```

```python
class MdUpdate:
    # Отвечает за виды того что к нам приходит в потоке данных
    def __init__(self):
        pass
```

```python
class CandleUpdate(MdUpdate):
    def __init__(self, open, close, high, low, volume, ts):
        self.open = open
        self.close = close
        self.high = high
        self.low = low
        self.volume = volume
        self.ts = ts
```

```python
class NewsUpdate(MdUpdate):
    def __init__(self, text, ts):
        self.text = text
        self.ts = ts
```

```python
class DataFlow:
    # Отвечает за поток данных (данные отсортирваны по времени)
    def __init__(self, local_ts, data):
        
        assert len(local_ts) == len(data), "Length of local_ts and data must be equal."
        self.local_ts = local_ts
        self.data = data
        self._index = 0
```

```python
class CandleDataFlow(DataFlow):
    def __init__(self, local_ts, data):
        super().__init__(local_ts, data)
        self._index = 0
    
    def get_update(self):
        if self._index >= len(self.local_ts):
            return None
        else:
            update = CandleUpdate(
                self.data['open'][self._index],
                self.data['close'][self._index],
                self.data['high'][self._index],
                self.data['low'][self._index],
                self.data['volume'][self._index],
                self.local_ts[self._index]
            )
            self._index += 1
            return update
```

##### Атрибуты самой стратегии 

```python
ticker = 'sber'
board = 'tqbr'

start_interval = '2020-01-01'
end_interval = '2023-11-01'
period = 'D' #'1m', '10m', '1h', 'D', 'W', 'M', 'Q' 

cash = 10000

position = 0
fee = 0.0004

# window_size = 10
strategy=MeanReversion(10, fee)
```

### Начинаю мэтчить данные

```python
##### ticker+board -> figi for Tinkoff API

def get_share_attributes(TOKEN=token, ticker='SBER', board='TQBR'):
    
    ticker = ticker.upper()
    board = board.upper()

    with Client(TOKEN, target=INVEST_GRPC_API_SANDBOX) as client:

        instr = client.instruments.share_by(id_type=2, class_code=board, id=ticker)
        
        itg = []

        itg.append(instr.instrument.figi )
        itg.append(instr.instrument.isin )
        itg.append(instr.instrument.ticker )
        itg.append(instr.instrument.class_code )
        itg.append(instr.instrument.lot )
        itg.append(instr.instrument.currency )
        
        return itg
        
get_share_attributes(TOKEN=token, ticker=ticker, board=board)
#figi ----- isin ----- ticker ----- board ----- lot ----- currency
```

```python
##### get AlgoPack historical data

#algoPack
def get_shares_candles(secs=[], start = '2020-01-01', end = '2023-11-01', period='D'): 

    x = pd.DataFrame()

    for i in secs:
        try:
            ticker = Ticker(i.upper())
            y = ticker.candles(date=start, till_date=end, period=period)
            y['ticker'] = i
            x = pd.concat([x , y], axis=0)
        except:
            pass
    return x

get_shares_candles(secs=[ticker]
                   , start = start_interval
                   , end = end_interval
                   , period=period).head()
```

```python
##### Back Test on AlgoPack Historical Data

# window_size = 10
strategy=MeanReversion(10, fee)

# ticker = 'sber'
# board = 'tqbr'

# start_interval = '2020-01-01'
# end_interval = '2023-11-01'
# period = 'D' #'1m', '10m', '1h', 'D', 'W', 'M', 'Q' 

# cash = 10000

# position = 0
# fee = 0.0004

# # window_size = 10
# strategy=MeanReversion(10, fee)

def strat_back_test(strategy=strategy
                   , ticker=ticker
                   , board=board
                    
                   , start=start_interval
                   , end=end_interval
                   , period=period
                   
                   , cash=cash
                   , position=position
                   , fee=fee
                   
                   , plot=False):
    
    
    
    df = get_shares_candles(secs=[ticker]
                   , start = start_interval
                   , end = end_interval
                   , period=period)
    
    #strategy = MeanReversion(10, 0.0004)

    
    logs = []
    shares_quant = []
    pnl_logs = []
    
    
    for i, row in df.iterrows():
        update = {'price': row['close']}
        delta_position = strategy.get_update(update)
        position += delta_position

        shares_quant.append(position)
        logs.append(delta_position)

        if delta_position == 0:
            pass
        elif delta_position > 0:
            cash -= delta_position*(row['close']*(1+strategy.fees))
        else:
            cash += delta_position*(row['close']*(1-strategy.fees))

        pnl_logs.append(cash+position*row['close'])
        
    
    meta = { 'data': df
        , 'basic': {'ticker': ticker
                      , 'board':board}
            , 'strategy_params': {'start': start
                                  , 'end': end
                                  , 'period': period}
            , 'portfolio_params': {'cash': cash
                                   , 'position': position
                                   , 'fee': fee}
            , 'itg_strategy_meta': {'our_trades': logs
                                    , 'quant_change': shares_quant
                                    ,'cash_change': pnl_logs }
           }
    
    if plot==True:
        plt.plot(pnl_logs)
        
    return meta
```

```python
meta = strat_back_test(strategy=strategy
                   , ticker=ticker
                   , board=board
                    
                   , start=start_interval
                   , end=end_interval
                   , period=period
                   
                   , cash=cash
                   , position=position
                   , fee=fee
                   
                   , plot=True)

meta.keys()
```
